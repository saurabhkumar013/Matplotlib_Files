```python
df_google_play_store_apps = pd.read_csv("C:\\Users\\Saurabh\\OneDrive\\Documents\\archive\\googleplaystore.csv", nrows = 1000)
df_google_play_store_apps.shape # numpy
```

```python
df_google_play_store_apps
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x,y, c = "r")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5,linewidths = 10)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5,linewidths = 10, edgecolors= "g")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5,linewidths = 10, edgecolors= "g", verts = "<") 
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", s = 100, alpha = 0.5,linewidths = 10, edgecolors= "g", verts = "<")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5,linewidths = 10, edgecolors= "g")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```
